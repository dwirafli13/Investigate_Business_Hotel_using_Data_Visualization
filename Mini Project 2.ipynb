# Data Preprocessing

```python
data.isna().sum()
```

## Mengatasi Data Null

```python
data_prep = data.copy()
```

```python
data_prep.drop(columns=['agent','company'], inplace=True)
```

```python
data_prep['city'] = data_prep['city'].fillna('unknown')
```

```python
data_prep['children'] = data_prep['children'].fillna(0)
```

```python
data_prep.isna().sum()
```

Langkah - langkah yang diambil :
- Menghapus kolom `agent` dan `company` karena terlalu banyak data yang null
- Mengisi nilai null pada kolom `city` dengan value `unknown` karena tidak diketahui asal kota dari customer
- Menganggap nilai null pada kolom `children` berarti tidak ada tamu anak kecil. Oleh karena itu diisi dengan nilai `0(Zero)`  

## Mengganti Value Yang Tidak Sesuai

```python
data_prep['meal'].unique()
```

```python
data_prep['meal'] = data_prep['meal'].replace(['Undefined'],'No Meal')
```

```python
data_prep['meal'].unique()
```

Mengganti value _Undefined_ menjadi _No Meal_ karena memiliki arti yang sama

## Membuang Data Yang Tidak Diperlukan

```python
data_prep['guest'] = data_prep['adults'] + data_prep['children'] + data_prep['babies']
```

```python
data_prep['guest'].unique()
```

```python
data_prep = data_prep[data_prep['guest'] > 0]
```

```python
data_prep['guest'].unique()
```

Menghapus data dengan jumlah tamu _0(zero)_

# Analisis Jumlah Pemesanan Hotel Setiap Bulan Berdasarkan Jenis Hotel

```python
data1 = data_prep.copy()
```

```python
data1_gb = data1.groupby(['hotel','arrival_date_month'])['arrival_date_year'].agg(['nunique','count']).reset_index()
data1_gb.columns = ['hotel_type','arrival_month','unique_year','total_booking']
data1_gb
```

Untuk data _October_ dan _September_ diambil dari data 3 tahun lalu, tetapi bulan lain diambil dari data 2 tahun lalu. Perlu dilakukan normalisasi data agar perbandingannya sesuai

```python
# Melakukan normalisasi data dengan menghitung rata-rata jumlah pesanan di suatu bulan pada tahun yang berbeda
data1_gb['rata_rata_total_booking'] = round(data1_gb['total_booking']/data1_gb['unique_year'])
data1_gb.head()
```

```python
ordered_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
data1_gb['month'] = data1_gb['arrival_month'].str[:3]
data1_gb['month'] = pd.Categorical(data1_gb['month'], categories=ordered_months, ordered=True)

data1_gb.head()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='month', y='rata_rata_total_booking', hue='hotel_type',
             size='hotel_type', sizes=(2.5, 2.5), data=data1_gb)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Tipe Hotel', title_fontsize=14, prop={'size':12})

plt.xlabel('Bulan Kedatangan', fontsize=15)
plt.ylabel('Rata Rata Total Booking', fontsize=15)
plt.ylim(0,4800)

plt.text(x=-0.5, y=5250, s='Kedua Tipe Hotel Memiliki Lebih Banyak Pengunjung Pada Musim Liburan',
        fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=4870, s='City Hotel Mengalami Penurunan Jumlah Pengunjung Yang Drastis Di Bulan Agustus Ke September, \nSedangkan Kedua Hotel Memiliki Lebih Sedikit Pelanggan Selama Bukan Musim Liburan (Jan-Mar)',
        fontsize=17)

plt.axvline(4, ls='--', color='green')
plt.axvline(6, ls='--', color='green')
plt.stackplot(np.arange(4,7,1),[[4800]], color='green', alpha=0.3)
plt.text(x=4.5, y=4400, s='Musim \nLiburan', fontsize=16, color='blue')

plt.axvline(9, ls='--', color='green')
plt.axvline(11, ls='--', color='green')
plt.stackplot(np.arange(9,12,1),[[4800]], color='green', alpha=0.3)
plt.text(x=9.5, y=4400, s='Musim \nLiburan', fontsize=16, color='blue')

plt.tight_layout()
plt.savefig('visualisasi_task_1.png', dpi=200)
```

# Analisis Pengaruh Durasi Menginap Terhadap Tingkat Pembatalan Pemesanan Hotel

```python
data2 = data_prep.copy()
```

```python
data2['total_nights'] = data2['stays_in_weekend_nights'] + data2['stays_in_weekdays_nights']
```

```python
data2.head()
```

```python
sns.histplot(data=data2, x='total_nights', binwidth=5)
```

Berdasarkan pada visualisasi di atas, dapat dilihat bahwa persebaran datanya terlihat right-skewed. Oleh karena itu akan dilakukan pengelompokkan agar persebarannya lebih merata. Pengelompokkan tersebut dilakukan untuk total jumlah malam yang mencapai 3 minggu maka dikelompokkan dari malam ke-15 sampai malam ke-21 (15-21). Selanjutnya untuk total malam yang lebih dari 3 minggu (>21)

```python
data2['total_nights_group'] = np.where(data2['total_nights'] > 21, 21, 
                                       np.where(data2['total_nights'] > 14, 15, data2['total_nights']))
```

```python
data2.head()
```

```python
data2_gb = data2.groupby(['hotel','is_canceled','total_nights_group']).agg({'arrival_date_month' : 'count'}).reset_index()
data2_pivot = data2_gb.pivot_table(index=['hotel', 'total_nights_group'],
                                  columns=['is_canceled'],
                                  values=['arrival_date_month']).reset_index()
data2_pivot.columns = ['hotel','total_nights_group','0','1']
data2_pivot['pct_canceled'] = round(data2_pivot['1']*100.0/(data2_pivot['0']+data2_pivot['1']),2)
```

```python
data2_pivot
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='total_nights_group', y='pct_canceled', hue='hotel', data=data2_pivot)
sns.regplot(x='total_nights_group', y='pct_canceled', data=data2_pivot[data2_pivot['hotel'] == 'City Hotel'], scatter=False, label='Trend City Hotel')
sns.regplot(x='total_nights_group', y='pct_canceled', data=data2_pivot[data2_pivot['hotel'] == 'Resort Hotel'], scatter=False, label='Trend Resort Hotel')

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Tipe Hotel', title_fontsize=15, prop={'size':13})

plt.xticks(range(0,17,1), labels=[str(x) for x in range(0,15,1)]+['15-21','>21'])
plt.xlabel('Total Malam', fontsize=15)
plt.ylabel('Persentase Pembatalan Booking (%)', fontsize=15)
plt.ylim(0, 100)
plt.xlim(0.5, 16.5)

plt.text(x=0.5, y=110, s='Tren Positif Pada Total Malam Dan Persentase Dari Pembatalan Pesanan', 
         fontsize=20, fontweight='bold')
plt.text(x=0.5, y=102, s='Semakin Lama Total Malam Yang Dipesan Maka Persentase Pembatalan Juga Akan Semakin Tinggi, \nLalu City Hotel Memiliki Tren Yang Lebih Curam Dibandingkan Dengan Resort Hotel', 
         fontsize=18)

plt.tight_layout()
plt.savefig('visualisasi_task_2.png', dpi=200)
```

# Analisis Pengaruh Jarak Waktu Pemesanan Terhadap Tingkat Pembatalan Pemesanan Hotel

```python
data3 = data_prep.copy()
```

```python
sns.histplot(data=data3, x='lead_time', bins=50)
```

```python
data3['lead_time'].max()
```

```python
data3['lead_times'] = np.where(data3['lead_time'] > 400, 401, data3['lead_time'])
bins = list(np.arange(0,max(data3['lead_times'])+30, 30))
labels = ["{0}-{1}".format(str(bins[x]), str(bins[x+1])) for x in range(len(bins[:-1]))]
data3['lead_time_group'] = pd.cut(data3['lead_times'],bins=bins, include_lowest=True, labels=labels)
```

```python
data3.head()
```

```python
data3_gb = data3.groupby(['hotel','lead_time_group','is_canceled']).agg({'lead_times' : 'count'}).reset_index()
data3_pivot = data3_gb.pivot_table(index=['hotel','lead_time_group'],
                                  columns=['is_canceled'],
                                  values=['lead_times']).reset_index()
data3_pivot.columns = ['hotel_type','lead_time_group','not_canceled', 'canceled']
data3_pivot['total_booking'] = data3_pivot['not_canceled']+data3_pivot['canceled']
data3_pivot['pct_canceled'] = round(data3_pivot['canceled']*100.0/(data3_pivot['total_booking']),2)
```

```python
data3_pivot.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='lead_time_group', y='pct_canceled', hue='hotel_type', data=data3_pivot)

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid()
plt.legend(title='Tipe Hotel', title_fontsize=15, prop={'size':13}, loc='upper center')

plt.xlabel('Lead Time Groups (per 30 hari)', fontsize=15)
plt.ylabel('Persentase Pembatalan Booking (%)', fontsize=15)
plt.ylim(0,100)
plt.xlim(-0.5,13.5)

plt.axvline(0.5, ls='--', color='green')
plt.stackplot(np.arange(-0.5,1.5,1),[[100]], color='green', alpha=0.3)
plt.text(x=-0.1, y=80, s='1 Bulan', fontsize=20, color='blue', rotation=90)

plt.axvline(8.5, ls='--', color='green')
plt.stackplot(np.arange(8.5,14.5,1),[[100]], color='green', alpha=0.3)
plt.text(x=10, y=90, s='Sekitar 1 Tahun', fontsize=20, color='blue')

plt.text(x=-0.5, y=110, s='Kedua Hotel Memiliki Rasio Pembatalan Paling Rendah Dalam 1 Bulan Lead Time', 
         fontsize=20, fontweight='bold')
plt.text(x=-0.5, y=102, s='Resort Hotel Memiliki Rasio Pembatalan Yang Cukup Stagnan (sekitar 40%), Sedangkan City Hotel \nMemiliki Rasio Pembatalan Yang Cukup Tinggi (di atas 60%) Dengan Lead Time Sekitar 1 Tahun', 
         fontsize=18)

plt.tight_layout()
plt.savefig('visualisasi_task_3.png', dpi=200)
```
